```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df= pd.read_csv(DATOS_DIR + 'Semillas.csv')
seleccion = df.iloc[:,0:-1]
seleccion
```

```python
plt.figure()
sns.heatmap(seleccion.corr(), annot=True)
plt.show()
```

```python
seleccion = seleccion.replace([np.inf, -np.inf], np.nan)
plt.figure()
sns.pairplot(seleccion)
plt.show()
```

```python
media = seleccion.mean()
desvio = seleccion.std()
seleccionNorm= (seleccion-media)/desvio
#seleccionNorm.sort_values('Area', ascending=False, inplace=True)
round(seleccionNorm,3)
```
