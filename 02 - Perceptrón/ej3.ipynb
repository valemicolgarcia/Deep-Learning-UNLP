La Tabla 1 muestra información correspondiente de pacientes para determinar si deben realizarse un 
examen médico en función de su edad, altura y riesgo médico.

Donde:
- EDAD es un atributo numérico que indica la edad del 
paciente.
- RIESGO es el nivel de riesgo del paciente.
- EXAMEN indica si debe realizarse un examen extra.

Para obtener transformar el atributo nominal RIESGO en uno 
numérico se lo numerizó de la siguiente forma: BAJO = 1, MEDIO = 2 y 
ALTO = 3.

1. Luego de la numerización se calculó el coeficiente de correlación lineal entre los atributos EDAD y RIESGO 
y se obtuvo como resultado -0.71. ¿Cómo debe interpretarse este valor?

Un valor de -0.71 indica una CORRELACION LINEAL NEGATIVA. Esto significa que a medida que EDAD aumenta, RIESGO disminuye y viceversa

2. Luego de numerizar el atributo RIESGO y de normalizar los atributos de manera lineal entre 0 y 1, los 
ejemplos fueron utilizados para entrenar un perceptrón capaz de predecir correctamente el atributo 
EXAMEN. Los pesos obtenidos fueron los siguientes:

¿Cuál será la respuesta del perceptrón para los siguientes valores?


```python
#W_EDAD = 0.0807
#W_RIESGO = 0.074
#B = -0.0742

```

```python
W_EDAD = -0.076
W_RIESGO = -0.071
B = 0.071
```

```python
BAJO = 1
MEDIO = 2
ALTO = 3
```

```python
EDAD_min = 53
EDAD_max = 60
def normalizar_edad(edad):
    edad_normalizada = (edad - EDAD_min) / (EDAD_max - EDAD_min)
    return edad_normalizada

```

```python
def normalizar_riesgo(riesgo):
    riesgo_normalizado = (riesgo - 1) / (3 - 1)
    return riesgo_normalizado
```

```python
def respuesta_perceptron (edad, riesgo):
    respuesta = 1
    y = W_EDAD * normalizar_edad(edad) + W_RIESGO * normalizar_riesgo(riesgo) + B
    respuesta = 1 if y>=0 else 0
    retorno = 'SI' if respuesta == 0 else 'NO'
    return retorno
```

```python
print(respuesta_perceptron(55,ALTO))
print(respuesta_perceptron(56,MEDIO))
print(respuesta_perceptron(58,MEDIO))
print(respuesta_perceptron(56,BAJO))
print(respuesta_perceptron(59,BAJO))
print(respuesta_perceptron(57,MEDIO))
print(respuesta_perceptron(60,BAJO))
print(respuesta_perceptron(53,ALTO))
print(respuesta_perceptron(59,MEDIO))

```
