Matriz de confusion

```python
confusion_matrix_data = [
    [17, 0, 1, 0, 1],
    [0, 12, 0, 0, 0],
    [0, 0, 12, 0, 0],
    [2, 0, 0, 38, 0],
    [0, 8, 0, 0, 61]
]
```

```python
true_labels = [0]*19 + [1]*12 + [2]*12 + [3]*40 + [4]*69
#19 ejemplos de la clase 0
#12 ejemplos de la clase 1
#12 ejemplos de la clase 2
#40 ejemplos de la clase 3
#69 ejemplos de la clase 4
```

```python
pred_labels = [0]*17 + [2] + [4] + [1]*12 + [2]*12 + [0]*2 + [3]*38 + [1]*8 + [4]*61
#clase 0: 17 clasificados como 0 bien, 1 clasificado como 2 mal, 1 clasificado como 4 mal
#clase 1: 12 bien clasificados como 1
#clase 2: 12 bien clasificados como 2
#clase 3: 38 bien clasificados como 3, 2 mal clasificados como 0
#clase 4: 61 bien clasificados como 4, 8 mal clasificados como 4
```

```python
print(len(true_labels))
print(len(pred_labels))
```

152 ejemplos de entrenamiento

5 clases diferentes 

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
```

```python
accuracy = accuracy_score(true_labels, pred_labels)
precision = precision_score(true_labels, pred_labels, average=None)
recall = recall_score(true_labels, pred_labels, average=None)
f1 = f1_score(true_labels, pred_labels, average=None)
```

### Accuracy

```python
print("Accuracy (Precisión Global):", accuracy)
```

El accuracy indica que el modelo tiene una precision global del 92.1%, del total de las predicciones realizadas, solo el 92.1% fueron correctas

### Precision

```python
print("Precision (Precisión por clase):", precision)
```

- Clase 0: el 89% de las veces que el modelo predice que un ejemplo pertenece a la clase 0, realmente es de clase 0
- Clase 1: Solo el 60% de las veces que dice clase 1 es clase 1 por lo que tiende a generar falsos positivos
- Clase 2: El 92% de las veces que el modelo predice que un ejemplo pertenece a la clase 2 es realmente de clase 2
- Clase 3: Todas las veces que el modelo predice que un ejemplo es de la clase 3, es de clase 3
- Clase 4 : El 98% de las veces que dice que es de clase 4, verdaderamente lo es. Lo podemos ver ya que solo 1 ejemplo fue mal clasificado como de clase 4

### Recall

```python
print("Recall (Sensibilidad por clase):", recall)
```

- Clase 0: El 89% de los ejemplos que realmente pertencen a la clase 0 se detectaron correctamente.
- Clase 1: El modelo detecto el 100% de los ejemplos que pertencen a la clase 1
- Clase 2: El modelo detecto el 100% de los ejemplos que pertenecen a la clase 1
- Clase 3: El modelo detecto el 95% de los ejemplos que pertencen a la clase 3
- Clase 4: El modelo detecto el 88.4% de los ejemplos que pertencen a la clase 4

### F1-Score

```python
print("F1-Score por clase-:", f1)
```

Las clases 0, 2, 3 y 4 tienen valores de F1-Score muy altos (entre 0.93 y 0.97), lo que indica que el modelo hace un buen trabajo al clasificar estos ejemplos. La clase 1 tiene el valor de F1-Score más bajo (0.75), lo que sugiere que el modelo tiene más problemas para clasificar correctamente los ejemplos de esta clase.
